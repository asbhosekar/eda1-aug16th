```python
'''
Question ->
Try to plot the pie chart for top 3 countries having
the highest number of restaurants in the given dataset.
'''
```

```python
# Solution ->
country_names = final_df['Country'].value_counts().sort_values(ascending = False)[:3].index
country_values = final_df['Country'].value_counts().sort_values(ascending = False)[:3].values

# Plot
plt.figure(figsize = (7, 7))
plt.pie(country_values, labels = country_names, autopct = '%.2f%%', shadow = True)
plt.title('Top 3 countries having highest number of restaurants')
plt.legend()
plt.show()
```

```python
top_countries = final_df['Country'].value_counts().sort_values(ascending=False)[:3]
country_names = top_countries.index
country_values = top_countries.values

# Step 2: Plot the pie chart
plt.figure(figsize=(6, 6))
explode = [0.1, 0, 0]  # Highlight the top country
plt.pie(
    country_values,
    labels=country_names,
    autopct='%.2f%%',
    shadow=True,
    startangle=140,
    explode=explode
)
plt.title('Top 3 Countries with Highest Number of Restaurants')
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Filter and group the data
white_rating_df = final_df[final_df['Rating color'] == 'White'] \
    .groupby('Country') \
    .size() \
    .reset_index(name='Frequency')

# Step 2: Extract labels and values
labels = white_rating_df['Country']
sizes = white_rating_df['Frequency']

# Optional: explode the largest slice
explode = [0.1 if freq == max(sizes) else 0 for freq in sizes]

# Step 3: Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Proportion of 0 Ratings by Country (Rating Color: White)')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Filter and group the data
online_delivery_df = final_df[final_df['Has Online delivery'] == 'Yes'] \
    .groupby('Country') \
    .size() \
    .reset_index(name='Frequency')

# Step 2: Extract labels and values
labels = online_delivery_df['Country']
sizes = online_delivery_df['Frequency']

# Optional: explode the largest slice (India)
explode = [0.1 if freq == max(sizes) else 0 for freq in sizes]

# Step 3: Plot the pie chart
plt.figure(figsize=(8, 6))
colors = plt.cm.plasma([0.2, 0.6])  # Custom colors for 2 slices
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=140, colors=colors)
plt.title('Countries Focusing on Online Deliveries')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.legend(labels, title="Country", loc="upper right")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Data
data = {'Country': ['India', 'UAE'], 'Frequency': [2423, 28]}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Country', y='Frequency', data=df, palette='viridis')
bar_plot.set_title('Frequency of Online Deliveries by Country', fontsize=14)
bar_plot.set_xlabel('Country', fontsize=12)
bar_plot.set_ylabel('Frequency', fontsize=12)

# Annotate bars
for index, row in df.iterrows():
    bar_plot.text(index, row.Frequency + 50, row.Frequency, color='black', ha="center", fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace with your actual DataFrame if needed)
data = {
    'Aggregate rating': [0.0, 0.0, 0.0, 0.0, 1.8],
    'Country': ['Brazil', 'India', 'United Kingdom', 'United States', 'India'],
    'Frequency': [5, 2139, 1, 3, 1]
}
df = pd.DataFrame(data)

# Set plot style
sns.set(style="whitegrid")

# Create clustered bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(
    x='Aggregate rating',
    y='Frequency',
    hue='Country',
    data=df,
    palette='viridis'
)

# Customize plot
bar_plot.set_title('Aggregate Rating Wise Country Wise Frequency', fontsize=14)
bar_plot.set_xlabel('Aggregate Rating', fontsize=12)
bar_plot.set_ylabel('Frequency', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

# Annotate bars
for p in bar_plot.patches:
    height = p.get_height()
    if height > 0:
        bar_plot.annotate(f'{int(height)}',
                          (p.get_x() + p.get_width() / 2., height),
                          ha='center', va='bottom',
                          fontsize=10, color='black', xytext=(0, 5),
                          textcoords='offset points')

plt.tight_layout()
plt.show()
```
